```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

```python
D = 1
v = 0.1
k = 0.0005
L = 1000
j0= -1
```

# The equation
$$
    \begin{cases}
    &Dy'' - vy' - ky = 0\\
    &Dy'(0) - vy(0) = j0\\
    &Dy'(L) - vy(L) = 0
    \end{cases}
$$

or with the solver

\begin{equation}
    \begin{cases}
        &y' = z \\
        &z' = ky + vz
    \end{cases}
\end{equation}

```python
def f(t,y):
    return np.array([y[1], k*y[0] + v*y[1]])

def bc(y0, yL):
    return np.array([D*y0[1] - v*y0[0] - j0,
                     D*yL[1] - v*yL[0]])

x = np.linspace(0,L,1000)
yi = np.zeros((2,x.shape[0]))

yi[0,500] = 1
```

```python
res = solve_bvp(f, bc, x, yi)

x_plot = np.linspace(0,L,100)
y_plot = res.sol(x_plot)[0]

fig, ax = plt.subplots(1,1)
ax.plot(x_plot, y_plot)
ax.set_ylim(0, y_plot.max()*1.1);
```
