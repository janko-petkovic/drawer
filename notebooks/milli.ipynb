```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p_diff = 0.01

# Generate male trajectories
rand_delta = np.random.rand(10000,1000)
male_delta = np.where(rand_delta < 0.5 - p_diff/2, 0, 1)
female_delta = np.where(rand_delta < 0.5 + p_diff/2, 0, 1)

male_end = male_delta.sum(axis=1)
female_end = female_delta.sum(axis=1)

avg_mf_diff = male_end.mean() - female_end.mean()
print(avg_mf_diff/female_end.mean())



fig, axs = plt.subplots(1,1)

ax = axs
mvals, mbins, _ = ax.hist(male_end, bins=30, density=False, alpha=0.5);
fvals, fbins, _ = ax.hist(female_end, bins=30, density=False, alpha=0.5);

mvals = mvals/10000
fvals = fvals/10000
```

```python
p_f_over_m = 0

for mv, mb in zip(mvals, mbins[:-1]):
    p_f_over_m += fvals[np.where(fbins[:-1]>=mb)].sum()*mv

print(p_f_over_m)
```
