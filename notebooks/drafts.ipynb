```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def null_v(v,tau,I):
    return tau * v + I

def null_w(v):
    return 6*v


v = np.linspace(-1,1,100)

fig, axs = plt.subplots(1,3,figsize=(7.5,2), dpi=200)
fig.subplots_adjust(wspace=0.5)

ax = axs[0]
w_null_v = null_v(v, 1, -0.2)
w_null_w = null_w(v)
ax.plot(v, w_null_v, label='w nullcline')
ax.plot(v, w_null_w, label='v nullcline')

ax.legend(ncols=2, loc=(1,-.6), frameon=False)
ax.set_title('Regular case', pad=10)

ax = axs[1]
w_null_v = null_v(v, 6, 2)
w_null_w = null_w(v)
ax.plot(v, w_null_v)
ax.plot(v, w_null_w)
ax.set_title(r'$\tau=6, \, I \neq 0$', pad=10)

ax = axs[2]
w_null_v = null_v(v, 6,0)
w_null_w = null_w(v)
ax.plot(v, w_null_v, linewidth=3)
ax.plot(v, w_null_w)

ax.set_title(r'$\tau=6, \, I = 0$', pad=10)

for ax in axs:
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
    ax.set_xticks([0])
    ax.set_yticks([0])
    ax.set_xlabel('v')
    ax.set_ylabel('w')
```
