```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
def f(x):
    y = 0
    for x_bar in [0.,2.,4.,6.,8.]:
        y += torch.exp(-(x-x_bar)**2/0.01)  
    
    return y
```

```python
x = torch.linspace(0,10,1000)
x.requires_grad = True

y = f(x)
y.sum().backward()

# dydx = x.grad
# x.grad.data.zero_()
# dydx.sum().backward()


```

```python
fig, axs = plt.subplots(2,1,figsize=(6,3),sharex=True)

ax = axs[0]
ax.plot(x.detach().numpy(), f(x).detach().numpy())
ax.set_aspect('equal')
ax.set_ylim(-0.1, 1.1)

ax = axs[1]
ax.plot(x.detach().numpy(), x.grad.detach().numpy())
# ax.set_aspect('equal')
```

```python
z = y.sum()
z.backward()

plt.plot(x.detach().numpy(), x.grad.detach().numpy())
ax = plt.gca()

ax.set_ylim(-0.1, 1.1)
```
